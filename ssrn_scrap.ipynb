## Classification Model 

```python
df1 = pd.read_csv('ml_abs.csv')
df2 = pd.read_csv('dl_abs.csv')
df3 = pd.read_csv('rf_abs.csv')
```

```python
data = pd.concat([df1,df2,df3],axis=0,ignore_index=True)
```

```python
data
```

```python
data['abstract'] = data['abstract'].str.split('Abstract\n',expand=True)[1]
```

```python
data.flag = data.flag.astype("category").cat.codes
```

```python
data['flag'] = data.flag.astype('category')
```

```python
data.flag.value_counts()
```

```python
data.flag.value_counts()
```

```python
## Converting to a function
def cleanText(text:str)-> str:
    """
    This function takes a tweet and clean it by removing
    1. Hastags and Mentions
    2. URLS
    """
    text = text.lower()
    hash_pattern  = re.findall("[#|@]\w+",text)
    for p in hash_pattern:
        text = text.replace(p,"")
    urls = re.findall("http[a-z:/0-9.]+",text)
    for u in urls:
        text = text.replace(u,"")
    text =re.findall("\w+",text)
    text = " ".join(text)
    return text
```

```python
data['abstract'] = data['abstract'].apply(cleanText)
```

```python
df = data.sample(frac=1).reset_index(drop=True)
```

```python
df.to_csv('ssrn_abs.csv')
```

```python
X = data.abstract
y = data.flag
```

```python
vectorizer = TfidfVectorizer(ngram_range=(1,3),min_df = 0.2)
vectorizer.fit(X)
X_vec = vectorizer.transform(X)
```

```python
X_vec
```

```python
x_train,x_test,y_train,y_test = train_test_split(X_vec,y,test_size=0.2,random_state=69)
```

```python
def objectivel(trial,x_train,y_train):
    c = trial.suggest_float('C',0,100)
    model = LinearSVC(C=c)
    score = cross_val_score(model,x_train,y_train,n_jobs=-1,cv=5,scoring='f1_macro')
    return score.mean()

study1 = optuna.create_study(direction='maximize',study_name = 'text_classflinear')
func = lambda trial: objectivel(trial,x_train,y_train)
study1.optimize(func,n_trials=100,n_jobs=-1)
```

## SVC

```python
def objective(trial,x_train,y_train):
    c = trial.suggest_float('C',0,100)
    k = trial.suggest_categorical('kernel',['linear', 'poly', 'rbf', 'sigmoid'])
    model = SVC(C=c,kernel=k)
    score = cross_val_score(model,x_train,y_train,n_jobs=-1,cv=5,scoring='f1_macro')
    return score.mean()
```

```python
study2 = optuna.create_study(direction='maximize',study_name = 'text_classf')
func = lambda trial: objective(trial,x_train,y_train)
study2.optimize(func,n_trials=100,n_jobs=-1)
```
